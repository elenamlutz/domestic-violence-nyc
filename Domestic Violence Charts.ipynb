```python
import pandas as pd
import requests
import json
import re
from bs4 import BeautifulSoup
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from pywaffle import Waffle
```

```python
df = pd.read_csv("male_race.csv")
```

```python
df2 = pd.read_csv("female_race.csv")
```

```python
colors = ['#8ac2b2', '#01505f']

ax = df.plot(x='Victim', y=['pct_population', 'pct_homicide'], color=colors, kind='bar', figsize=(12,4))

ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_facecolor('#CACACA')
ax.tick_params(axis='x', rotation=360)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.savefig('male_race.pdf')

ax.tick_params(which='both', left=False, bottom=False, grid_alpha=0.9)
```

```python
data = [['Weapon', 25], ['Physical Attack', 5]] 
df = pd.DataFrame(data, columns = ['Method', 'Number']) 
```

```python
df['pct'] = ((df['Number'] * 100) / 3000) * 100
```

```python
matplotlib.rcParams['pdf.fonttype'] = 42 #makes text editable in illustrator
%matplotlib inline
fig = plt.figure(
    FigureClass=Waffle, 
    rows=5,
    colors=("#0f4c76", "white"),
    title={'label': "Use of Weapons Have Increased", "loc": "left"},
    values=df.pct,
    labels=list(df.Method),
    figsize=(13, 10),
    font_size=32,
    legend={'loc': 'upper left', 'bbox_to_anchor': (1.1, 1)}
)
plt.tight_layout()
plt.savefig('weapons.pdf')
```
